# 04 Business Insights

This notebook translates analysis results into business insights and recommendations.

```python
from pathlib import Path

import pandas as pd

PROJECT_ROOT = Path.cwd().parents[0] if Path.cwd().name == 'notebooks' else Path.cwd()
DATA_PATH = PROJECT_ROOT / 'data' / 'processed' / 'crop_production_cleaned.csv'

df = pd.read_csv(DATA_PATH)
latest_year = df['year'].max()
latest = df[df['year'] == latest_year].copy()
latest[['crop', 'production_tonnes', 'production_share_pct', 'volatility_score']].sort_values('production_tonnes', ascending=False)
```

```python
top_crop = latest.sort_values('production_tonnes', ascending=False).iloc[0]
high_volatility_crop = latest.sort_values('volatility_score', ascending=False).iloc[0]

print(f"Latest year: {latest_year}")
print(f"Top crop by production: {top_crop['crop']}")
print(f"Highest volatility crop: {high_volatility_crop['crop']}")
```
