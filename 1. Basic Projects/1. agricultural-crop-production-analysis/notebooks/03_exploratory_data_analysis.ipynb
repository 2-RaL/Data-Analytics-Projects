# 03 Exploratory Data Analysis

This notebook analyzes production, area, yield, growth, and volatility.

```python
from pathlib import Path

import matplotlib.pyplot as plt
import pandas as pd

PROJECT_ROOT = Path.cwd().parents[0] if Path.cwd().name == 'notebooks' else Path.cwd()
DATA_PATH = PROJECT_ROOT / 'data' / 'processed' / 'crop_production_cleaned.csv'

df = pd.read_csv(DATA_PATH)
df.head()
```

```python
yearly = df.groupby('year', as_index=False).agg(
    total_production_tonnes=('production_tonnes', 'sum'),
    total_area_harvested_ha=('area_harvested_ha', 'sum')
)
yearly['weighted_yield_kg_per_ha'] = yearly['total_production_tonnes'] * 1000 / yearly['total_area_harvested_ha']
yearly
```

```python
latest_year = df['year'].max()
latest = df[df['year'] == latest_year].sort_values('production_tonnes', ascending=False)
latest[['crop', 'production_tonnes', 'area_harvested_ha', 'yield_kg_per_ha', 'production_share_pct']].head(10)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(yearly['year'], yearly['total_production_tonnes'] / 1000, marker='o')
plt.title('Total Crop Production Trend')
plt.xlabel('Year')
plt.ylabel('Production (thousand tonnes)')
plt.tight_layout()
plt.show()
```

```python
volatility = df.groupby('crop', as_index=False)['production_yoy_growth_pct'].std()
volatility = volatility.rename(columns={'production_yoy_growth_pct': 'volatility_score'})
volatility.sort_values('volatility_score', ascending=False)
```
